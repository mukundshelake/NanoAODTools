```python
import uproot
import numpy as np

file = uproot.open(filePath)
tree = file['Events']
```

```python
for key in tree.keys():
    for k in ['MuonIDWeight', 'MuonHLTWeight', 'LHEWeightSign', 'bTaggingWeight', 'jetPUIdWeight']:
        if k in key:
            array = tree[key].array(library='np')
            max_val = np.max(array)
            min_val = np.min(array)
            mean_val = np.mean(array)
            std_val = np.std(array)
            print(f"{key}: max={max_val}, min={min_val}, mean={mean_val}, std={std_val}")
```
